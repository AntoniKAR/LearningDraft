```python
# –í–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã. –ß–∞—Å—Ç—å 2.
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ a –∏ b (a < b). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ 
# —á–∏—Å–ª–æ –∏–∑ –æ—Ç—Ä–µ–∑–∫–∞ [a;b] —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π –¥–µ–ª–∏—Ç–µ–ª–µ–π.
a , b = int(input()), int(input())
total_maximum = 0
digit = 0

for i in range(a, b + 1):
    maximum = 0
    for j in range(1, i + 1):
        if i % j == 0:
            maximum += j
        if maximum >= total_maximum:
            total_maximum = maximum
            digit = i
print(digit, total_maximum)
```

```python
# –î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n.
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å –≤—ã—Å–æ—Ç–æ–π —Ä–∞–≤–Ω–æ–π n, –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–æ–º:
n=int(input())
total=0
for i in range (1, n+1):
    for j in range (1,i+1):
        print(1+total, end=" ")
        total+=1
    print()
```

```python
# –î–∞–Ω–æ –Ω–µ—á–µ—Ç–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. 
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —Ä–∞–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–Ω—ã–π –∑–≤–µ–∑–¥–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º, —Ä–∞–≤–Ω—ã–º nn –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–æ–º:
n=int(input())
for i in range(1,n+1):
    while 1<=i<=((n//2)+1):
        for j in range (i):
            print("*", end="")
        print()
        break
    while ((n//2)+1)<i<=n:
        for j in range ((n-i)+1):
            print("*", end="")
        print()
        break
        
```

```python
# –†–∞—Å–∫–æ–¥–∏—Ä–æ–≤–Ω–∏–µ —à–∏—Ñ—Ä–∞ —Ü–µ–∑–∞—Ä—è

n=int(input())
w=input()
for s in w:
    if ord(s)-n>=97:
        print(chr(ord(s)-n), end="")
    else:
        print(chr(122-(n-(ord(s)-97)-1)), end="")                                                                                                                                                                                                                                                                                                                                                                                                                   
```

```python
# –°—É–º–º—ã –¥–≤—É—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n –≥ge n‚â•2, –∞ –∑–∞—Ç–µ–º n —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª.
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –∏–∑ —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —á–∏—Å–µ–ª —Å–ø–∏—Å–æ–∫, 
# —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ —Å—É–º–º —Å–æ—Å–µ–¥–Ω–∏—Ö —á–∏—Å–µ–ª (0 –∏ 1, 1 –∏ 2, 2 –∏ 3 –∏ —Ç.–¥.).

n, ls, lk = int(input()), [], []
for i in range(n):
    n1 = int(input())
    lk.append(n1)
for k in range(1,n):
    ls.append(lk[k] + lk[k-1])
print(ls)
```

```python
# GOOGLE SEARCH
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n, –∑–∞—Ç–µ–º n —Å—Ç—Ä–æ–∫, –∑–∞—Ç–µ–º —á–∏—Å–ª–æ k ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∏—Å–∫–æ–≤—ã—Ö –∑–∞–ø—Ä–æ—Å–æ–≤,
# –∑–∞—Ç–µ–º k —Å—Ç—Ä–æ–∫ ‚Äî –ø–æ–∏—Å–∫–æ–≤—ã–µ –∑–∞–ø—Ä–æ—Å—ã.–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤—Å–µ –≤–≤–µ–¥–µ–Ω–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏,
# –≤ –∫–æ—Ç–æ—Ä—ã—Ö –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –≤—Å–µ –ø–æ–∏—Å–∫–æ–≤—ã–µ –∑–∞–ø—Ä–æ—Å—ã.
n=int(input())
st=[]
for i in range(n):
    s=input()
    st.append(s)
k=int(input())
st1=[]
for m in range(k):
    s1=input()
    st1.append(s1)
for j in range(len(st)):
    count=0
    for g in range(len(st1)):
        if st1[g].lower() in st[j].lower():
            count+=1
        else:
            continue
    if count==k:
        print(st[j], sep='\n')
    
```

```python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–≤–ø–∞–¥–∞—é—â–∏—Ö –ø–∞—Ä

# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–µ —á–∏—Å–ª–∞. –ò–∑ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª. 
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –≤ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º —Å–ø–∏—Å–∫–µ –ø–∞—Ä —ç–ª–µ–º–µ–Ω—Ç–æ–≤, —Ä–∞–≤–Ω—ã—Ö –¥—Ä—É–≥ –¥—Ä—É–≥—É.
# –°—á–∏—Ç–∞–µ—Ç—Å—è, —á—Ç–æ –ª—é–±—ã–µ –¥–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–∞, —Ä–∞–≤–Ω—ã–µ –¥—Ä—É–≥ –¥—Ä—É–≥—É –æ–±—Ä–∞–∑—É—é—Ç –æ–¥–Ω—É –ø–∞—Ä—É, –∫–æ—Ç–æ—Ä—É—é –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å.

a = [int(s) for s in input().split()]
counter = 0
for i in range(len(a)):
    for j in range(i + 1, len(a)): # —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞ –∏ –ø–æ—Å–ª–µ–¥—É—é—â–∏–π, —Å –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∫–æ–π –ø—Ä–æ–º–µ–∂—É—Ç–∫–æ–≤: –≤ –ø–µ—Ä–≤–æ–º —Å–ª—É—á–∞–µ –æ—Ç 0                            # –¥–æ 
        if a[i] == a[j]:           # –¥–æ len(a), –≤–æ –≤—Ç–æ—Ä–æ–º –æ—Ç i+1(—Ç–æ–µ—Å—Ç—å –ø–æ—Å–ª–µ–¥—É—é—â–∏–π) –¥–æ len(a).
            counter += 1
print(counter)
```

```python
# –°–û–†–¢–ò–†–û–í–ö–ê –°–õ–ò–Ø–ù–ò–ï–ú!
# —Å–Ω–∞—á–∞–ª–∞ –ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ —É–∂–µ –¥–≤—É—Ö –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö! —Å–ø–∏—Å–∫–æ–≤, –∑–∞—Ç–µ–º —Ä–∞—Å—à–∏—Ä—è–µ–º –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –Ω–∞ –ª—é–±–æ–µ –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
# –ø—Ä–∏ –ø–æ–º–æ—â–∏ –≤–≤–µ–¥–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ quick_merge2. –ö–æ—Ç–æ—Ä–∞—è –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç —Å–ª–∏—è–Ω–∏–µ —É–∂–µ –ª—é–±–æ–≥–æ –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–ø–∏—Å–∫–æ–≤! 

def quick_merge(list1, list2):
        q1=0
        q2=0
        result=[]
        while q1<len(list1) and q2<len(list2):
            if list1[q1]<list2[q2]:
                result.append(list1[q1])
                q1+=1
            else:
                result.append(list2[q2])
                q2+=1
        if q1<len(list1):
            result+=list1[q1:]
        if q2<len(list2):
            result+=list2[q2:]
        return result
    
def quick_merge2():
    total_list=[]
    for i in range(int(input())):
        num=[int(k) for k in input().split()]
        total_list=quick_merge(total_list, num)
    return total_list

quick_merge2()
```

```python
# –§–£–ù–ö–¶–ò–Ø –ü–ï–†–ï–í–û–î–ê –ò–ó –í–ï–†–ë–õ–Æ–ñ–ï–ì–û –†–ï–ì–ò–°–¢–†–ê –í –ó–ú–ï–ò–ù–´–ô

def convert_to_python_case(text):
    ls=[]
    pos=[]
    for i in range(len(text)):
        ls.append(text[i])
        if text[i] in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
            pos.append(i)
    del pos[0]
    for k in range(len(pos)):
        ls.insert(pos[k]+k, "_")
    t="".join(ls)
    return t.lower()

# —Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ
txt = input()
# –≤—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é
print(convert_to_python_case(txt))
```

```python
# –ü–∞–Ω–≥—Ä–∞–º–º—ã
#–ü–∞–Ω–≥—Ä–∞–º–º–∞ ‚Äì —ç—Ç–æ —Ñ—Ä–∞–∑–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è –≤ —Å–µ–±–µ –≤—Å–µ –±—É–∫–≤—ã –∞–ª—Ñ–∞–≤–∏—Ç–∞.
#–û–±—ã—á–Ω–æ –ø–∞–Ω–≥—Ä–∞–º–º—ã –∏—Å–ø–æ–ª—å–∑—É—é—Ç –¥–ª—è –ø—Ä–µ–∑–µ–Ω—Ç–∞—Ü–∏–∏ —à—Ä–∏—Ñ—Ç–æ–≤, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ –≤ –æ–¥–Ω–æ–π —Ñ—Ä–∞–∑–µ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –≤—Å–µ –≥–ª–∏—Ñ—ã.

#–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, is_pangram(text) –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ —Å—Ç—Ä–æ–∫—É —Ç–µ–∫—Å—Ç–∞
#–Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º —è–∑—ã–∫–µ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ True –µ—Å–ª–∏ —Ç–µ–∫—Å—Ç —è–≤–ª—è–µ—Ç—Å—è –ø–∞–Ω–≥—Ä–∞–º–º–æ–π –∏ False –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.
# –æ–±—ä—è–≤–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
def is_pangram(text):
    ls=[]
    ls.extend(text)
    ls_new=[]
    for i in range(len(ls)):
        if ls[i]!=" ":
            ls_new.append(ls[i])
    new_str="".join(ls_new).lower()
    if len(set(new_str))==26: # —Ñ—É–Ω—Ü–∏—è set –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —ç–ª–º–µ–Ω—Ç–æ–≤ —Å—Ç—Ä–æ–∫–∏(–≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è —É–¥–∏–Ω–æ–∂—ç–¥—ã –≤ —Å—Ç—Ä–æ–∫–µ) 
        return True
    else:
        return False
# —Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ
text = input()

# –≤—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é
print(is_pangram(text))
```

```python
# quicksort –∞–ª–≥–æ—Ä–∏—Ç–º
from random import *
def quicksort(num):
    if len(num) <= 1:
        return num
    else:
        p = choice(num)
        l_row = [i for i in num if i < p]
        e_row = [p] * num.count(p)
        r_row = [k for k in num if k > p]
        return quicksort(l_row) + e_row + quicksort(r_row)

k = [1, 3, 4, 10, 1, 12, 123, 342, 234, 2356, 1, 124, 334, 12123, 232, 353, 54657, 7578, 212, 7, 756, 12, 424, 123124, 6767]
quicksort(k)
    
        
```
