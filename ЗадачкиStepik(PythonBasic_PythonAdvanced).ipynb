```python
# Вложенные циклы. Часть 2.
# На вход программе подается два натуральных числа a и b (a < b). Напишите программу, которая находит натуральное 
# число из отрезка [a;b] с максимальной суммой делителей.
a , b = int(input()), int(input())
total_maximum = 0
digit = 0

for i in range(a, b + 1):
    maximum = 0
    for j in range(1, i + 1):
        if i % j == 0:
            maximum += j
        if maximum >= total_maximum:
            total_maximum = maximum
            digit = i
print(digit, total_maximum)
```

```python
# СОРТИРОВКА СЛИЯНИЕМ!
# сначала пишем функцию для сортировки уже двух отсортированных! списков, затем расширяем возможности на любое колличество сортировки
# при помощи введнения функции quick_merge2. Которая осуществляет слияние уже любого колличества отсортированных списков! 

def quick_merge(list1, list2):
        q1=0
        q2=0
        result=[]
        while q1<len(list1) and q2<len(list2):
            if list1[q1]<list2[q2]:
                result.append(list1[q1])
                q1+=1
            else:
                result.append(list2[q2])
                q2+=1
        if q1<len(list1):
            result+=list1[q1:]
        if q2<len(list2):
            result+=list2[q2:]
        return result
    
def quick_merge2():
    total_list=[]
    for i in range(int(input())):
        num=[int(k) for k in input().split()]
        total_list=quick_merge(total_list, num)
    return total_list

quick_merge2()
```

```python
# ФУНКЦИЯ ПЕРЕВОДА ИЗ ВЕРБЛЮЖЕГО РЕГИСТРА В ЗМЕИНЫЙ

def convert_to_python_case(text):
    ls=[]
    pos=[]
    for i in range(len(text)):
        ls.append(text[i])
        if text[i] in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
            pos.append(i)
    del pos[0]
    for k in range(len(pos)):
        ls.insert(pos[k]+k, "_")
    t="".join(ls)
    return t.lower()

# считываем данные
txt = input()
# вызываем функцию
print(convert_to_python_case(txt))
```

```python
# Панграммы
#Панграмма – это фраза, содержащая в себе все буквы алфавита.
#Обычно панграммы используют для презентации шрифтов, чтобы можно было в одной фразе рассмотреть все глифы.

#Напишите функцию, is_pangram(text) которая принимает в качестве аргумента строку текста
#на английском языке и возвращает значение True если текст является панграммой и False в противном случае.
# объявление функции
def is_pangram(text):
    ls=[]
    ls.extend(text)
    ls_new=[]
    for i in range(len(ls)):
        if ls[i]!=" ":
            ls_new.append(ls[i])
    new_str="".join(ls_new).lower()
    if len(set(new_str))==26: # фунция set возвращает множество уникальных элментов строки(встречающихся удиножэды в строке) 
        return True
    else:
        return False
# считываем данные
text = input()

# вызываем функцию
print(is_pangram(text))
```

```python
# Функция перевода из 10 системы счисления в любую другую
# ответом является распакованный список *

def from_ten_to_any(num, new_base):
    dig=[]
    while num//new_base>=1:
        dig.append(num%((num//new_base)*new_base))
        num=num//new_base
    dig+=[num]
    dig.reverse()
    return print(*dig, sep=" ") # *

num=int(input())
new_base=int(input())

from_ten_to_any(num,new_base)

# * ответом является распакованный список 
```

```python
# Функция перевода 2.0 из 10-ой системы счисления в любую другую
# ответом является распакованный список*
def from_ten_to_any(num, new_base):
    dig=[]
    while num//new_base>=1:
        dig.append(num%((num//new_base)*new_base))
        num=num//new_base
    dig+=[num]
    dig.reverse()
    if new_base==16:
        for e in dig:
            if e>=10:
                letter=chr(e+55)
                dig[dig.index(e)]=letter
    return print(*dig, sep="")

num=int(input())
new_base=int(input())

from_ten_to_any(num,new_base)

# усоверенствованна тем, что выводи по "стандарту" число в шестнадцатеричной системе исчисления
```

```python
# Функция set() под капотом
# Высчитывает колличесвто уникальных переменных в списке
def back_stage_set(objects):
    count=0
    unique_set=[]
    for i in range(len(objects)):
        for j in range(len(objects)):
            if objects[i] is objects[-len(objects)+j]:
                count+=1
        if count==1:
            unique_set.append(objects[i])
        elif count>1 and (objects[i] not in unique_set):
            unique_set.append(objects[i])
    return (len(unique_set))
objects = [1,1,1,1,1,1,2,15.8,15,15,13,16,18]
back_stage_set(objects)
```

```python
# пример рекурсии для нахождения колличества сочетаний из n элементов по k в группе
n,k=map(int, input().split())
def soch(n,k):
    if k>n:
        return 0
    elif k==0:
        return 1
    else:
        return soch(n-1,k)+soch(n-1,k-1)

print(soch(n,k))
```

```python
# Эмулятор пространства имен

n = int(input())

parent = {"global": None}
vs = {"global": set()}

for _ in range(n):
    t, fst, snd = input().split()
    if t == "create":
        parent[fst] = snd
        vs[fst] = set()
    elif t == "add":
        vs[fst].add(snd)
        print(vs)
    else:  # t == get
        while fst is not None:
            if snd in vs[fst]:
                break
            fst = parent[fst]
        print(fst)
```

```python
# Задача Флавия: из n человек в кругу умирает каждый k- тый
n=int(input())
k=int(input())
def flav(n,k): # n- колличество человек в круге, k- выбывает каждый k-тый
    if n == 1:
        return 1
    else:
        return (1+((flav(n-1,k)+k-1) % n))
flav(n,k)
```

```python
# quicksort алгоритм
from random import *
def quicksort(num):
    if len(num) <= 1:
        return num
    else:
        p = choice(num)
        l_row = [i for i in num if i < p]
        e_row = [p] * num.count(p)
        r_row = [k for k in num if k > p]
        return quicksort(l_row) + e_row + quicksort(r_row)

k = [1, 3, 4, 10, 1, 12, 123, 342, 234, 2356, 1, 124, 334, 12123, 232, 353, 54657, 7578, 212, 7, 756, 12, 424, 123124, 6767]
quicksort(k)
    
        
```

```python
# треугольник паскаля его n-ная строчка
n=int(input())
ls=[[1],[1,1]]
for i in range(2 ,n+1):
    ls.append([])
    for m in range(i+1):
        if m == 0 or m == i:
            ls[i].append(1)
        else:
            ls[i].append(ls[i-1][m-1]+ls[i-1][m])
print(ls[n])
```
