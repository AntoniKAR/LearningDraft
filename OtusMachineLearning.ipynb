###  Генерация случайных величин Рандом в NumPy



```python
import random
# напоминалка модуля random из стандартного пайтона
# random.uniform(2,3) выдает тслучайные числа с плавующей точко в диапазоне [2,3)
unif = random.uniform(2,3)
print(unif)
print()
# random.random() выдает тслучайные числа с плавующей точко строго в диапазоне [0,1)
ran = random.random()
print(ran)
print()
# random.randint(1,9) выдает тслучайное число в диапазоне [1,9] (диапазон может быть любой)
randit = random.randint(1,9)
print(randit)
print()
# Randonm в списках
# random.shufle(a), где a - список, на выходе получим премешанный список а, причем изменяет изначальный список, а не создает копию 
a = [1,2,3,4,5,6]
random.shuffle(a)
print(a)
print()
# random.choice(a) возвращает случайный элемент из списка
print(random.choice(a))
print()
# random.sample(a) тоже самое что и choice, но есть возможность вывести несколько случайных элементов из списка !тоже в виде списка!
print(random.sample(a ,2))
print()

# -------------------------------------------------------------------------------------------------

# np.random.rand(4,4), данный метод аналогичен питоновскому uniform(0,1), работающей на промежутке [0,1), или random(),
# всегда рабтающему на промежутке [0,1)
print("Получили матрицу 4 на 3 из рандомных чисел на промежутке от 0 включительно до 1 невключительно")
uni_rand = np.random.rand(4,3)
print(uni_rand)
print()
# Для того чтобы поменять промежуток нам необходимо премножить 5*np.random.rand(4) например на пять тем самым получим аналогию
# uniform(0,5) в стандартном питоне
print("Получили матрицу 15 на 2 из рандомных чисел на промежутке от 0 включительно до 5 невключительно")
uni_rand_multi = 5 * np.random.rand(15,2) # 15, 2 - размерность матрицы
print(uni_rand_multi)
print()
# или двумерной...
uni_rand_dimens = 5 * np.random.rand(6,6)
print("Получили матрицу 6 на 6 из рандомных чисел на промежутке от 0 включительно до 5 невключительно")
print(uni_rand_dimens)
print()
# Генерация массива с из целых чисел с помощью np.random.randint(0, 10, 10) где 0 - левая граница диапазона включительно,
#  10- правая гранциа включительно, и вторая 10 - колличество элементов в списке
print("Получили массив из 3 случайных целых чисел из диапазона от 0 до 10 включительно")
arr_int = np.random.randint(0, 10, size = 3)
print(arr_int) # Получили массив из 3 случайных целых чисел из диапазона от 0 до 10 включительно: [3 1 0]
print()
print()
    
#  Используя функцию seed можем зафиксировать начальный вектор генерации псевдослучайных чисел, тоесть при неизменном
#  количестве рандомны чисел в массиве всегда будет генерироваться один и тот же набор данных в массиве, а вот с увеличением
#  количесвта чисел в массиве будет прибавляться и новые рандомные числа в зависимости от изначального знчения seed
#  тоесть seed это как бы начальное значение для генерации рандома от которого отталкивается генератор
import numpy as np
from numpy.random import RandomState, SeedSequence, MT19937
rs = RandomState(MT19937(SeedSequence(16))) # меняя этот параметр с 16 на любое другое число будут меняться и данные в массиве
print("Получили фиксированный набор данных в масиве, меняя seed изменим и набор")
print(rs.randint(1, 10, 3))
# так же можем просто задать seed изначально без привязки метода к переменной
np.random.seed(12)
print(np.random.randint(1,10,3))
print()
# Удобная функция NumPy филтр-маска..   Данн массив num_array = np.array([1,2,3,4,5,6,10,15,15,23,23,12,756,34,21,346,2,673,34])
num_array = np.array([1,2,3,4,5,6,10,15,15,23,23,12,756,34,21,346,2,673,34])
print(num_array[num_array > 10]) # Получили данные из массива которые больше 10
print()
mask = num_array > 10
print("массив из True или False в зависимости удровлетворяет ли элемент массива условию или нет")
print(mask)   # полчуили массив в котором на метсах соотвествующих оригинальному массиву написанно True если параметр
print()       # удовлетворяет заданному условию  > 10 или False если параметр не удовлетворяет заданному условию 
              # так же можно накладывать нескольок условия с помощью логических опреаций
mask_two = (num_array > 10) & (num_array < 23)
print("Получили отфильтрованный массив согласно улсовию")
print(num_array[mask_two])
print(mask_two) 





```

```python
# посчтиаем среднее wage в зависимости от female
print("получили среднее по male и female")
print(df_w.groupby("female")["wage"].mean())
print()
# тоже самое, нов более длинном виде
print("тоже самое, но в более длинном виде")
print(df_w.wage[df_w.female == 1].mean(), df_w.wage[df_w.female == 0].mean(), end = ' ')
print()

# Более большое условие группровки доавим женаты или нет
print("Более большое условие группровки доавим женаты или нет")
print(df_w.groupby(["female", "married"])["wage"].mean())
print()
# Так жек можем привести в болле читабельную форму добавив unstack()
print("Так жек можем привести в болле читабельную форму добавив unstack()")
print(df_w.groupby(["female", "married"])["wage"].mean().unstack())
print()
# ил в жругом виде добавив в качестве аргумента as_index = False к методу groupby
print("или в жругом виде добавив в качестве аргумента as_index = False к методу groupby")
print(df_w.groupby(["female", "married"], as_index = False)["wage"].mean())
```
